# Handling Missing Values

```python
import numpy as np
import pandas as pd
path='C:/Users/HP 450 G8/Downloads/archive (6)/disney_plus_titles.csv'
Disney=pd.read_csv(path)
```

```python
#set seed for reproducibility
np.random.seed(0)
```

# Take a first look at the data

```python
Disney.head()
```

# how many missing data points do we have?

```python
missing_values_count=Disney.isnull().sum()
```

```python
missing_values_count[0:10]
```

# How many total missing values do we have ?

```python
total_cells=np.product(Disney.shape)
total_missing=missing_values_count.sum()
percent_missing=(total_missing/total_cells)*100
print(percent_missing)
```

# Drop missing values

```python
Disney.dropna()
```

```python
columns_with_na_dropped=Disney.dropna(axis=1)
columns_with_na_dropped.head()
```

# How much data did we lose ?

```python
print("Columns in original dataset :%d \n" %Disney.shape[1])
print("Columns with na's dropped:%d"%columns_with_na_dropped.shape[1])
```

# Filling in missing values

```python
Disney.fillna(0)
```

```python
Disney.fillna(method='bfill',axis=0).fillna(0)
```
